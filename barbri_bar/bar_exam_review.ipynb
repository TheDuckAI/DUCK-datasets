```python
print(len(p['Problem Number'].unique()))
print(len(p['Problem Number']))
p = p[p['Problem Number'] != "NAN"]
print(len(p['Problem Number'].unique()))
print(len(p['Problem Number']))
p = p.drop_duplicates(subset=['Problem Number'], keep='first')
print(len(p['Problem Number'].unique()))
print(len(p['Problem Number']))
p.reset_index(drop=True)
p = p.reset_index(drop=True)
```

```python
def create_category(x):
    return re.sub(r' ?(\d+) ?', r'', x)
```

```python
p['Category'] = p.apply(lambda row : create_category(row['Problem Number']), axis = 1)
```
